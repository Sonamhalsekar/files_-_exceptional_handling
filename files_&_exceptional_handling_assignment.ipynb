Q1. What is the difference between interpreted and compiled languages?
  - Compiled languages translate the whole code into machine language before running, making them faster. Examples include C and C++. Interpreted languages run the code line by line, which is slower but easier to test and debug. Examples are Python and JavaScript. Compiled ones are better for speed, while interpreted ones are more flexible.
